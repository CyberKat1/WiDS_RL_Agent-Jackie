```python
import numpy as np
```

```python
num_states, num_actions = 0, 0

with open('./data/continuing-mdp-50-20.txt','r') as f:
    for lines in f:
        l=lines.split()
        if l[0]=='numStates':
            num_states=int(l[1])
        elif l[0]=='numActions':
            num_actions=int(l[1])
            rewards = np.zeros((num_states, num_actions, num_states))
            probabilities = np.zeros((num_states, num_actions, num_states))
        elif l[0]=='end':
            terminal_states = []
            for _ in l[1:]:
                terminal_states.append(int(_))
        elif l[0]=='transition':
            s1 = int(l[1])
            a = int(l[2])
            s2 = int(l[3])
            r = float(l[4])
            p = float(l[5])

            rewards[s1][a][s2] = r
            probabilities[s1][a][s2] = p
        elif l[0]=='discount':
            gamma=float(l[1])

# print(probabilities)
# print(rewards)
print('discount factor =',gamma)
print('terminal states =',terminal_states)
```

```python
i=0
e=1e-6
t=True
mdp = np.zeros((num_states, num_actions, num_states))
state_values = np.zeros(num_states)
optimal_states = np.zeros(num_states)
optimal_actions = np.zeros(num_states)

while t==True:
    for s in range(num_states):
        for a in range(num_actions):
            # mdp[s][a] = probabilities[s][a] * (rewards[s][a] + gamma * state_values)
            mdp[s][a] = rewards[s][a] @ probabilities[s][a] + gamma * (probabilities[s][a] @ state_values)
        i+=1
        print('******************************\n',mdp[s])
        best_state_value = np.max(mdp[s])
        optimal_states[s] = best_state_value
        # optimal_actions[s]=np.argmax(mdp[s])
        optimal_actions[s]=np.argmax(np.max(mdp[s],axis=1))
        print(best_state_value)
        print(np.argmax(np.max(mdp[s],axis=1)))
        if abs(state_values[s] - optimal_states[s]) < e and (s not in terminal_states):
            t=False
            break
        state_values[s] = optimal_states[s]

print('number of iterations = '+str(i))
print('best states')
print(optimal_states)
print('best actions')
print(optimal_actions)
```

```python
with open('solution-continuing-50-20.txt','w') as f:
    for _ in range(len(optimal_states)):
        f.write(str(optimal_states[_])+' '+str(int(optimal_actions[_]))+'\n')
        print(str(optimal_states[_]),str(int(optimal_actions[_])))
```
